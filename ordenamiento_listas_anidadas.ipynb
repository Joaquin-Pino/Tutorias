verificar que un sudoku sea valido, por ahora, solo verificar que filas y columnas sean validas

```python
# vamos a verificar solo las filas y columnas, las submatrices las verificaremos cuando veamos funciones propias
# aqui se define el tablero de sudoku como una lista de listas
# cada lista interna representa una fila del tablero
sudoku = [
    [5, 3, 4, 6, 7, 8, 9, 1, 2],
    [6, 7, 2, 1, 9, 5, 3, 4, 8],
    [1, 9, 8, 3, 4, 2, 5, 6, 7],
    [8, 5, 9, 7, 6, 1, 4, 2, 3],
    [4, 2, 6, 8, 5, 3, 7, 9, 1],
    [7, 1, 3, 9, 2, 4, 8, 5, 6],
    [9, 6, 1, 5, 3, 7, 2, 8, 4],
    [2, 8, 7, 4, 1, 9, 6, 3, 5],
    [3, 4, 5, 2, 8, 6, 1, 7, 9]
]

# se inicia una variable para rastrear si el sudoku es correcto
# empezamos asumiendo que es valido y si encontramos un error la cambiamos a falso
valido = True

# ------ verificamos las filas------
# este primer ciclo 'for' recorre cada fila del tablero, una por una
for i in range(len(sudoku)):
    # para cada fila, creamos una lista vacia para guardar los numeros que ya hemos visto
    vistos = []
    # este segundo ciclo 'for' recorre cada numero dentro de la fila actual
    for j in range(len(sudoku[0])):
        # guardamos el numero de la celda actual en una variable
        numero = sudoku[i][j]
        # revisamos si el numero actual ya esta en nuestra lista de 'vistos'
        if numero in vistos:
            # si el numero ya estaba, significa que esta repetido en la misma fila
            # por lo tanto, el sudoku no es valido y cambiamos la variable a falso
            valido = False
        else:
            # si el numero no estaba, lo agregamos a la lista 'vistos'
            # para poder compararlo con los siguientes numeros de la fila
            vistos.append(numero)

# ------ verificamos las columnas ------------
# este ciclo 'for' recorre cada columna
# la variable 'i' ahora representa el indice de la columna que estamos revisando
for i in range(len(sudoku)):
    # de nuevo, creamos una lista vacia para los numeros vistos en la columna actual
    vistos = []
    # este segundo ciclo 'for' recorre las celdas de la columna actual
    # la variable 'j' ahora representa el indice de la fila
    for j in range(len(sudoku[0])):
        # para obtener el numero de la columna, invertimos los indices: sudoku[j][i]
        # asi nos movemos hacia abajo en lugar de hacia la derecha
        numero = sudoku[j][i]
        # revisamos si el numero ya esta en nuestra lista de 'vistos' de la columna
        if numero in vistos:
            # si el numero ya estaba, la columna tiene un repetido y el sudoku es invalido
            valido = False
        else:
            # si no estaba, lo agregamos a la lista de 'vistos' de esta columna
            vistos.append(numero)

```

```python
#entrada = input("ingrese secuencia de inputs")

entrada = "UUDDLLRRAUDLFUDLFADURLDURLADDUUDDUUALRLRUDLRUDLRUDDDURLD"
# inicializar las variables que usaremos para contar
batallas_ganadas = 0
caminata_actual = 0
caminata_mas_larga = 0
en_batalla = False # un "interruptor" para saber si estamos dentro de una batalla

# iterar sobre cada input en la secuencia
for caracter in entrada:
    
    # comprobar si el caracter es 'A', indica si se inicia o termina una batalla
    if caracter == 'A':
        # si NO estábamos en batalla, este 'A' inicia una
        if en_batalla == False:
            en_batalla = True
            
            # comparamos la caminata que acabamos de terminar
            if caminata_actual > caminata_mas_larga:
                caminata_mas_larga = caminata_actual
            
            # reiniciamos la caminata actual
            caminata_actual = 0
        
        # si YA estábamos en batalla, este 'A' la termina
        else:
            en_batalla = False
            batallas_ganadas += 1
   #si no es 'A', es un paso de movimiento (U, D, L, R)
    else:
        # solo contamos los pasos si NO estamos en batalla
        if en_batalla == False:
            caminata_actual += 1

# verificación final después del bucle
# la ultima caminata (si la hubo) debe ser comparada
if caminata_actual > caminata_mas_larga:
    caminata_mas_larga = caminata_actual

print(batallas_ganadas, caminata_mas_larga)
```

# das menu
ejercicio, mostrar este menu de menor a mayor precio

```python
# menu reducido, habria que hacer lo mismo para cada item del menu
# menu es representadoc como una lista donde cada elemento es otro lista con los datos de un plato
MENU = [["Salchichas con chucrut", 5000, ["gluten", "lactosa", "mostaza"]],
        ["Schnitzel", 6500, ["gluten", "huevo"]],
        ["Spatzle", 4500,  ["gluten", "huevo"]]]
# definicion de constantes
# asi usamos 'NOMBRE' en vez del numero 0 para referirnos al nombre del plato
NOMBRE = 0
PRECIO = 1
ALERGENOS = 2

# ------ entrada ------

tope_precio = int(input("ingrese tope precio" ))
alergico = input("¿Es usted alergico o alergica? (S/N) ")

# se prepara una lista vacia para guardar las alergias del usuario
alergias = []
# si la respuesta del usuario fue 's'
if alergico == "S":
    # se le pide que ingrese sus alergenos
    # el metodo 'split(",")' separa el texto por las comas y crea una lista con ellas
    alergias = input("Ingrese sus alergenos separados por coma: ").split(",")

# --------- seccion de procesamiento ------

# se crea una lista vacia para ir guardando los platos que cumplen los requisitos
platos_aceptados = []
# este ciclo 'for' revisa cada 'plato' que esta dentro de la lista 'MENU'
for plato in MENU:
    # se asume que el plato no da alergia al principio
    # esta variable 'bandera' se reinicia para cada plato
    hay_alergia = False
    # este ciclo 'for'  revisa cada 'alergeno' en la lista de alergenos del plato
    for alergeno in plato[ALERGENOS]:
        # se comprueba si el alergeno del plato esta en la lista de alergias del usuario
        if alergeno in alergias:
            # si lo esta, marcamos que este plato si causa alergia
            hay_alergia = True

    # se revisan las dos condiciones para aceptar un plato
    # la primera es si el precio se ajusta
    # la segunda es que la bandera 'hay_alergia' siga siendo falsa
    if tope_precio >= plato[PRECIO] and hay_alergia == False:
        # si se cumplen las condiciones, se agrega el plato completo a la lista de aceptados
        platos_aceptados.append(plato)

# ------ seccion de salida ------------------

# se recorre la lista que ahora solo contiene los platos que el usuario puede comer
for plato in platos_aceptados:
    # se imprime en pantalla cada plato aceptado
    print(plato)
```

# ordenamiento

```python

lista = [5,8,7,9,0,3,4,2,1,5]

# el primer ciclo 'for' recorre toda la lista
# la variable 'i' actua como un separador entre la parte ordenada (izquierda) y la no ordenada (derecha)
for i in range(len(lista)):
    # al comienzo de cada pasada, asumimos que el primer elemento de la parte no ordenada es el mas pequeño
    min_valor = lista[i]
    # y guardamos su posicion o indice
    min_indice = i
    
    # este segundo ciclo 'for' busca en el resto de la lista (la parte no ordenada)
    # para encontrar si hay un numero aun mas pequeño
    for j in range(i+1, len(lista)):
        # se compara cada numero de la parte no ordenada con el minimo valor que hemos encontrado hasta ahora
        if lista[j] < min_valor:
            # si encontramos un numero mas pequeño, lo actualizamos como el nuevo minimo
            min_valor = lista[j]
            # y tambien actualizamos la posicion de ese numero minimo
            min_indice = j

    # una vez que el segundo ciclo 'for' termina, ya sabemos cual es el numero mas pequeño
    # de la parte no ordenada y donde esta
    # ahora hacemos el intercambio (swap)
    
    # guardamos en una variable temporal el valor que esta en la posicion 'i'
    temp = lista[i]
    # ponemos el valor minimo que encontramos en la posicion 'i'
    lista[i] = min_valor
    # y en la posicion donde estaba el minimo, ponemos el valor que originalmente estaba en 'i'
    lista[min_indice] = temp

# despues de que el primer ciclo 'for' termina, la lista esta completamente ordenada
print(lista)
```

# como ordenar una lista de cosas?

```python
# aplicamos el mismo algoritmo de ordenamiento anterior, solo que ahora estamos tratando con una lista de estudiantes
# cada estudiante es una lista que tiene una id, nombre y nota

# definicinon de constantes
ID = 0
NOMBRE = 1
NOTA = 2

# --------- entrada ------------------------
# se pregunta cuantos estudiantes se van a registrar
numero_est = int(input("ingrese numero de estudiantes: "))

# ------------ procesamiento ---------
# primero, pedimos los datos de cada estudiante
# se crea una lista vacia principal para guardar a todos los estudiantes
estudiantes = []

# este ciclo se repite la cantidad de veces que indico el usuario
for i in range(0, numero_est):
    # para cada vuelta, se crea una lista temporal para el nuevo estudiante
    estudiante = []
    
    # se piden los datos 
    id_est = int(input("ingrse id: "))
    nombre = input("ingrese nombre: ")
    nota = float(input("ingrse nota: "))

    # se agregan los datos a la lista temporal 'estudiante'
    estudiante.append(id_est) #[1]
    estudiante.append(nombre) #[1, "nombre"]
    estudiante.append(nota)   # [1, "nombre", 5.0]

    # finalmente, se agrega la lista del estudiante a la lista principal 'estudiantes'
    estudiantes.append(estudiante)
    
#--------------- ahora ordenamos la lista de estudiantes segun la nota ------

# este es el algoritmo de ordenamiento por seleccion, adaptado para la lista de estudiantes
# el primer ciclo 'for' establece la posicion que vamos a ordenar ('i')
for i in range(0, numero_est):

    # asumimos que la nota mas baja es la del estudiante en la posicion 'i'
    nota_min = estudiantes[i][NOTA]
    # y guardamos el indice de ese estudiante
    indice_min = i
    
    # el segundo ciclo 'for' recorre el resto de la lista para buscar una nota mas baja
    for j in range(i + 1, numero_est):
        # si la nota minima que teniamos es mayor que la nota del estudiante en la posicion 'j'
        if nota_min > estudiantes[j][NOTA]:
            # encontramos una nueva nota minima, asi que la guardamos
            nota_min = estudiantes[j][NOTA]
            # y lo mas importante, guardamos el indice del estudiante con esa nota mas baja
            indice_min = j

    # al terminar el ciclo 'j', 'indice_min' apunta al estudiante con la menor nota de la seccion no ordenada
    # ahora hacemos el intercambio (swap) de los estudiantes completos
    temp = estudiantes[i]
    estudiantes[i] = estudiantes[indice_min]
    estudiantes[indice_min] = temp

# --- seccion de salida ---
# una vez que la lista esta ordenada por nota, la recorremos para mostrarla
for estudiante in estudiantes:
    # para cada estudiante en la lista ordenada, imprimimos su nombre y su nota
    print(estudiante[NOMBRE], "obtuvo", estudiante[NOTA])
```
