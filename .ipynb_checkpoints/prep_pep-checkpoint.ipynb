Ramujan es el l√≠der de un temible ej√©rcito de soldados, su ej√©rcito jam√°s ha conocido una
derrota en combate, sin embargo, ellos no son invencibles, pues cuando el fuego se interpone
entre ellos y su enemigo, simplemente escapan de la batalla, sin luchar.
Dadas estas condiciones, si tenemos la localizaci√≥n de los soldados de Ramujan, de los
soldados enemigos y del fuego, podemos determinar cu√°ntos soldados enemigos sobreviven.
Para ello, supongamos que los soldados del ej√©rcito de Ramujan se identificar√°n con ‚ÄòX‚Äô, sus
enemigos con ‚ÄòO‚Äô y el fuego con ‚Äò*‚Äô
.

Para determinar cu√°ntos enemigos sobreviven consideraremos que, cada vez que exista un
camino desde una ‚ÄòX‚Äô hasta una ‚ÄòO‚Äô, sin fuego de por medio, el soldado de Ramujan elimina
a su enemigo, sin importar si el soldado ‚ÄòX‚Äô est√° superado num√©ricamente o no. Sin embargo,
si existe fuego en el camino de ‚ÄòX‚Äô hacia ‚ÄòO‚Äô, los soldados enemigos sobreviven al ataque.

A fin de poder saber cu√°ntos enemigos sobrevivir√°n al terrible ej√©rcito de Ramujan, construya
una funci√≥n que reciba como entrada un string, representando el campo de batalla y entregue
la cantidad de soldados enemigos que sobreviven.

Por ejemplo:
‚Ä¢ ‚ÄòX*OO*XX‚Äô, resultar√≠a en 2 sobrevivientes, pues los dos soldados enemigos est√°n
protegidos por el fuego.

‚Ä¢ ‚ÄòX*OX**XO*XOX‚Äô, no tendr√≠a sobrevivientes, pues todos los soldados enemigos
tienen al menos un soldado de Ramujan adyacente a √©l.

‚Ä¢ ‚ÄòX*OO*OO*X*OX*‚Äô, tendr√≠a 4 sobrevivientes, pues existen 4 soldados protegidos por
el fuego.

Considere para su soluci√≥n que no es posible utilizar el m√©todo .count(), ni la palabra
reservada in

```python
campo_batalla = "X00*00X*000" # si no nos dicen que tipo de entrada es, eligan input(). en este caso pusimos la string para 
                                #que quede mas claro sobre lo que estamos trabajando
campo_batalla = campo_batalla.split("*")     # cada elemento de esta lista corresponde a sub batallas, debemos verificar si hay algo que
                                            # pueda matar a los soldados
total_sobrevivientes = 0 # si se termina de verificar una sub batalla y determinamos que no hay monstruo que mate a soldados, los sobrevivientes los contamos aca
enemigos_sub_batalla = 0 
hay_monstruo = False

i = 0
largo = len(campo_batalla)
while i < largo:
    sub_batalla = campo_batalla[i]
    # vamos a contar si hay algo que pueda matar a los soldados (O) en cada sub batalla
    hay_monstruo = False
    enemigos_en_sub_batalla = 0 # aun no hemos visto a nadie que sea no sea una X
    j = 0
    while j < len(sub_batalla):
        if sub_batalla[j] == '0':
            # encontramos un enemigo
            enemigos_en_sub_batalla = enemigos_en_sub_batalla + 1
        
        else: # es una X 
            # no importa si estaba antes o despues, ya puede matar a todos los soldados.
            # no contamos 
            hay_monstruo = True

        j+= 1

    #una vez terminada de revisar la sub batalla, preguntamos si se salvaron los soldados ( no hubo x)
    if hay_monstruo == False:
        total_sobrevivientes += enemigos_en_sub_batalla
    
        
    i += 1

    
print(total_sobrevivientes)
```

### Un antiguo hechicero dej√≥ en su tratado de magia ‚ÄúQu√¶ est in vita magicis‚Äù la teor√≠a de que
el poder de los hechizos reside en c√≥mo se utilizan las vocales para construirlo. Seg√∫n su
teor√≠a:

‚Ä¢ Un hechizo es pulentus si todas las vocales que se utilizaron en su construcci√≥n
aparecen en orden alfab√©tico.

‚Ä¢ Un hechizo es bacanus si las vocales que utiliza aparecen en orden alfab√©tico
inverso.

‚Ä¢ Un hechizo es cumast, si las vocales que utiliza no tienen en orden alguno.
Por ejemplo:

‚Ä¢ occidis simia, es bacanus pues las vocales que tiene est√°n en orden alfab√©tico
invertido.

‚Ä¢ modicum cattus, es cumast, pues las vocales no tienen orden alfab√©tico alguno.

‚Ä¢ sedivid oud, es pulentus, pues las vocales aparecen en orden alfab√©tico

Construya una funci√≥n en Python que reciba un string y clasifique el hechizo seg√∫n las
condiciones que cumpla, considere que la entrada siempre contendr√° s√≥lo letras en min√∫scula.

Considere que, si el hechizo tiene una sola letra, por ejemplo, vini, vidi, vici, cualquiera de las
categor√≠as ser√≠a correcta.

```python
# ESTUDIANDO MAGIA
hechizo = "sedivid oud"
referencia = "aeiou"

vocales_hechizo = []
# extraer las vocales
for letra in hechizo:
    if letra in referencia:
        vocales_hechizo.append(letra.lower())

# casos borde, estrictamente no son necesarios, ya que estos casos son ambiguos, puede ser pulentus o bacanus
if len(vocales_hechizo) == 0:
    print( "pulentus") 
if len(vocales_hechizo) == 1:
    print("pulentus")

es_pulentus = True
es_bacanus = True

# recorremos las vocales del hechizo
for i in range(len(vocales_hechizo) - 1):
    
    vocal_actual = vocales_hechizo[i]
    vocal_siguiente = vocales_hechizo[i+1]
    
    idx_actual = referencia.index(vocal_actual) # con esto miramos en nuesta lista de vocales (referencia)
    idx_siguiente = referencia.index(vocal_siguiente) 
    
    diferencia = idx_siguiente - idx_actual

    if diferencia < 0:
        es_pulentus = False
    
    if diferencia > 0:
        es_bacanus = False

# salida
if es_pulentus:
    print("pulentus")
elif es_bacanus:
    print("bacanus")
else:
    print("cumast")
```

### Johnny es un ling√ºista estudiando c√≥mo se utilizan las vocales en los distintos idiomas del
mundo, en particular, seg√∫n su teor√≠a, idiomas basados en el lat√≠n, como el espa√±ol, italiano y
franc√©s tienden a tener m√°s palabras con vocales consecutivas, mientras que otras lenguas
europeas tienen menos palabras con vocales consecutivas.

A fin de poder terminar su investigaci√≥n Johnny necesita un m√©todo r√°pido de determinar la
cantidad de vocales consecutivas dentro de una palabra, por lo que solicita a usted la
construcci√≥n de una funci√≥n en Python que automatice el proceso, es decir, se recibe como
entrada un string y se entrega como respuesta el largo del substring que contenga m√°s vocales
sin interrupciones, por ejemplo:

ÔÇ∑ Si la entrada fuese: ‚Äògroovy‚Äô, el substring m√°s largo de vocales es de largo 2, ‚Äòoo‚Äô

ÔÇ∑ Si la entrada fuese: ‚Äòoiseaux‚Äô, el substring m√°s largo de vocales es de largo 3, ‚Äòeau‚Äô

ÔÇ∑ Si la entrada fuese: ‚Äòhubschrauber‚Äô, el substring m√°s largo de vocales es de largo
2, ‚Äòau‚Äô

Considere que su entrada siempre estar√° compuesta solamente por caracteres alfab√©ticos sin
tilde y en min√∫scula. Para su soluci√≥n, no puede utilizar las palabras reservadas for e in.

```python
# TESIS DE VOCALES
palabra = "groovy"

contador_actual = 0
contador_maximo = 0
largo_palabra = len(palabra)

indice = 0
while indice < largo_palabra:
    
    letra = palabra[indice]
    es_vocal = False
    if letra == "a" or letra == "e" or letra == "i" or letra == "o" or letra == "u":
        es_vocal = True
    
    if es_vocal:
        # Si es vocal, aumentamos la racha actual
        contador_actual += 1
        
        if contador_actual > contador_maximo:
            contador_maximo = contador_actual
    else:
        contador_actual = 0
        
    indice += 1
    
contador_maximo
```

### Los n√∫meros de Armstrong son una de las muchas categor√≠as de n√∫meros creadas en la
matem√°tica recreacional, una rama de las matem√°ticas orientada a plantear juegos y puzles
matem√°ticos, adem√°s de problemas que alimentan la curiosidad y el deseo por resolver
problemas matem√°ticos complejos.

Se define un n√∫mero de Armstrong como un valor entero positivo de x d√≠gitos, que cumple
con la propiedad de que el n√∫mero es igual a la suma de los d√≠gitos elevados a la xava potencia.
Por ejemplo:

ÔÇ∑ 8 = 81 = 8

ÔÇ∑ 371 = 33 + 73 + 13 = 27 + 343 + 1 = 371

ÔÇ∑ 407 = 43 + 03 + 73 = 64 + 0 + 343 = 407

Construya un programa que dado un n√∫mero n entregue por pantalla el valor del navo n√∫mero
de Armstrong.

```python
#armstrong

numero = "407"
armstrong = False
prueba = 0
n = 0
while prueba < int(numero):
    prueba = 0
    for digito in numero:
        prueba += int(digito) ** n

    if prueba == int(numero):
        armstrong = True
    n += 1

armstrong
```

### Un n√∫mero fuerte es un n√∫mero que cumple la condici√≥n de que la suma de los factoriales de
sus d√≠gitos son iguales al n√∫mero original, por ejemplo:

‚Ä¢ 2 es un n√∫mero fuerte pues: 2! = 2

‚Ä¢ 145 es un n√∫mero fuerte pues: 1! + 4! + 5! = 1 + 24 + 120 = 145

‚Ä¢ 23 no es un n√∫mero fuerte pues: 2! + 3! = 2 + 6 = 8

Construya un programa en Python que permita determinar si un n√∫mero es fuerte o no.

Considere que para su soluci√≥n no puede utilizar funciones importadas.

```python
# numeros fuertes

numero = "23" 
suma = 0

for digito in numero:
    digito = int(digito)
    fact_digito = 1
    #calcular el factorial del digito
    for i in range(1, digito +1):
        fact_digito = fact_digito * i

    suma += fact_digito

if suma == int(numero):
    print("True")
else:
    print("False")
```

### Escritos antiguos en las paredes se√±alan que un n√∫mero de propiedades m√°gicas puede ser
encontrado utilizando la f√≥rmula:
ùë• = ùëõ2 + (ùëõ + 1)2
Siendo n un entero positivo y obteniendo el 85vo n√∫mero primo que esta genera. 
Se sabe que
esa f√≥rmula puede generar n√∫meros primos, pero tambi√©n se sabe que no todos los n√∫meros
que genera lo son. 
Por ejemplo, para n = 1, se genera 5, que es un n√∫mero primo, pero para
n = 3 genera 25, que no lo es. Escriba un programa que encuentre el 85vo n√∫mero primo
generado por esa f√≥rmula, para descifrar el misterio.

```python
# descifrando el misterio

cant_primos = 0
primo_misterioso = 0
n = 1
while cant_primos != 85:
    x = n**2 + (n + 1)**2

    # comprobar si x es primo
    primo = True
    for i in range(2, x):
        if x % i == 0:
            primo = False
            break # no ocupen esto, lo puse solo para que pudiera correr el algoritmo y mandar los apuntes

    if primo == True:
        cant_primos += 1
        primo_misterioso = x

    n += 1
primo_misterioso 
```
